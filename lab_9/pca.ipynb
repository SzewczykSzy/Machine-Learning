```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

iris = load_iris()
X = iris.data
y = iris.target

data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

M_cov = np.cov(X_std.T)
```

```python
from numpy.linalg import eig

w, v = eig(M_cov)
w
```

```python
z = X_std@v
```

```python
Z_1 = z[y[:]==0, 0]
Z_2 = z[y[:]==1, 0]
Z_3 = z[y[:]==2, 0]
X_std_1 = X_std[y[:]==0, 0]
X_std_2 = X_std[y[:]==1, 0]
X_std_3 = X_std[y[:]==2, 0]


fig, ax = plt.subplots(2,2, figsize=(15,15))

ax[0,0].scatter(Z_1, np.zeros(Z_1.shape[0]), marker='o', c='red')
ax[0,0].scatter(Z_2, np.zeros(Z_2.shape[0]), marker='x', c='green')
ax[0,0].scatter(Z_3, np.zeros(Z_3.shape[0]), marker='*', c='blue')
ax[0,0].legend(iris.target_names)
ax[0,0].set_xlabel('Z1')


ax[0,1].scatter(X_std_1, np.zeros(Z_1.shape[0]), marker='o', c='red')
ax[0,1].scatter(X_std_2, np.zeros(Z_2.shape[0]), marker='x', c='green')
ax[0,1].scatter(X_std_3, np.zeros(Z_3.shape[0]), marker='*', c='blue')
ax[0,1].legend(iris.target_names)
ax[0,1].set_xlabel('X1')


ax[1,0].scatter(Z_1, z[y[:]==0, 1], marker='o', c='red')
ax[1,0].scatter(Z_2, z[y[:]==1, 1], marker='x', c='green')
ax[1,0].scatter(Z_3, z[y[:]==2, 1], marker='*', c='blue')
ax[1,0].legend(iris.target_names)
ax[1,0].set_xlabel('Z1')
ax[1,0].set_ylabel('Z2')


ax[1,1].scatter(X_std_1, X_std[y[:]==0, 1], marker='o', c='red')
ax[1,1].scatter(X_std_2, X_std[y[:]==1, 1], marker='x', c='green')
ax[1,1].scatter(X_std_3, X_std[y[:]==2, 1], marker='*', c='blue')
ax[1,1].legend(iris.target_names)
ax[1,1].set_xlabel('X1')
ax[1,1].set_ylabel('X2')

plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,15), subplot_kw=dict(projection='3d'))

ax[0].scatter(Z_1, z[y[:]==0, 1], z[y[:]==0, 2], marker='o', c='blue')
ax[0].scatter(Z_2, z[y[:]==1, 1], z[y[:]==1, 2], marker='o', c='orange')
ax[0].scatter(Z_3, z[y[:]==2, 1], z[y[:]==2, 2], marker='o', c='green')
ax[0].legend(iris.target_names)
ax[0].set_xlabel('Z1')
ax[0].set_ylabel('Z2')
ax[0].set_zlabel('Z3')

ax[1].scatter(X_std_1, X_std[y[:]==0, 1], X_std[y[:]==0, 2], marker='o', c='blue')
ax[1].scatter(X_std_2, X_std[y[:]==1, 1], X_std[y[:]==1, 2], marker='o', c='orange')
ax[1].scatter(X_std_3, X_std[y[:]==2, 1], X_std[y[:]==2, 2], marker='o', c='green')
ax[1].legend(iris.target_names)
ax[1].set_xlabel('X1')
ax[1].set_ylabel('X2')
ax[1].set_zlabel('X3')
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('USArrests.csv', index_col=0)
df.head()
```

```python
df.info()
```

```python
df.mean()
```

```python
df.var()
```

```python
from sklearn.preprocessing import scale
X = pd.DataFrame(scale(df), index=df.index, columns=df.columns)
```

```python
from sklearn.decomposition import PCA
pca_loadings = pd.DataFrame(PCA().fit(X).components_.T, index=df.columns, columns=['V1', 'V2', 'V3', 'V4'])
pca_loadings
```

```python
pca = PCA()
df_plot = pd.DataFrame(pca.fit_transform(X), columns=['PC1', 'PC2', 'PC3', 'PC4'], index=X.index)
df_plot.head()
```

```python
fig , ax1 = plt.subplots(figsize=(9,7))
ax1.set_xlim(-3.5,3.5)
ax1.set_ylim(-3.5,3.5)
# Plot Principal Components 1 and 2
for i in df_plot.index:
    ax1.annotate(i, (-df_plot.PC1.loc[i], -df_plot.PC2.loc[i]), ha='center')
# Plot reference lines
ax1.hlines(0,-3.5,3.5, linestyles='dotted', colors='grey')
ax1.vlines(0,-3.5,3.5, linestyles='dotted', colors='grey')
ax1.set_xlabel('First Principal Component')
ax1.set_ylabel('Second Principal Component')
# Plot Principal Component loading vectors, using a second y-axis.
ax2 = ax1.twinx().twiny()
ax2.set_ylim(-1,1)
ax2.set_xlim(-1,1)
ax2.set_xlabel('Principal Component loading vectors', color='red')
# Plot labels for vectors. Variable 'a' is a small offset parameter to separate arrow tip and t
a = 1.07
for i in pca_loadings[['V1', 'V2']].index:
    ax2.annotate(i, (-pca_loadings.V1.loc[i]*a, -pca_loadings.V2.loc[i]*a), color='red')
# Plot vectors
ax2.arrow(0,0,-pca_loadings.V1[0], -pca_loadings.V2[0])
ax2.arrow(0,0,-pca_loadings.V1[1], -pca_loadings.V2[1])
ax2.arrow(0,0,-pca_loadings.V1[2], -pca_loadings.V2[2])
ax2.arrow(0,0,-pca_loadings.V1[3], -pca_loadings.V2[3])

plt.show()
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(7,5))
plt.plot([1,2,3,4], pca.explained_variance_ratio_, '-o')
plt.ylabel('Proportion of Variance Explained')
plt.xlabel('Principal Component')
plt.xlim(0.75,4.25)
plt.ylim(0,1.05)
plt.xticks([1,2,3,4])
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot([1,2,3,4], np.cumsum(pca.explained_variance_ratio_), '-s')
plt.ylabel('Proportion of Variance Explained')
plt.xlabel('Principal Component')
plt.xlim(0.75,4.25)
plt.ylim(0,1.05)
plt.xticks([1,2,3,4])
plt.show()
```

```python
df2 = pd.read_csv('NCI60.csv').drop('Unnamed: 0', axis=1)
df2.columns = np.arange(df2.columns.size)
df2.info()
```

```python
y = pd.read_csv('NCI60_y.csv', usecols=[1], skiprows=1, names=['type'])
```

```python
X = pd.DataFrame(scale(df2))
X.shape
```

```python
pca2 = PCA()
df2_plot = pd.DataFrame(pca2.fit_transform(X))
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,6))
color_idx = pd.factorize(y.type)[0]
cmap = mpl.cm.hsv
# Left plot
ax1.scatter(df2_plot.iloc[:,0], df2_plot.iloc[:,1], c=color_idx, cmap=cmap, alpha=0.5, s=50)
ax1.set_ylabel('Principal Component 2')
# Right plot
ax2.scatter(df2_plot.iloc[:,0], df2_plot.iloc[:,2], c=color_idx, cmap=cmap, alpha=0.5, s=50)
ax2.set_ylabel('Principal Component 3')
# Custom legend for the classes (y) since we do not create scatter plots per class (which could 
handles = []
labels = pd.factorize(y.type.unique())
norm = mpl.colors.Normalize(vmin=0.0, vmax=14.0)
for i, v in zip(labels[0], labels[1]):
    handles.append(mpl.patches.Patch(color=cmap(norm(i)), label=v, alpha=0.5))
ax2.legend(handles=handles, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
# xlabel for both plots
for ax in fig.axes:
    ax.set_xlabel('Principal Component 1')
```

```python
pd.DataFrame([df2_plot.iloc[:,:5].std(axis=0, ddof=0).to_numpy(), pca2.explained_variance_ratio_[:5], np.cumsum(pca2.explained_variance_ratio_[:5])],
            index=['Standard Deviation', 'Proportion of Variance', 'Cumulative Proportion'], columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
```

```python
df2_plot.iloc[:,:10].var(axis=0, ddof=0).plot(kind='bar', rot=0)
plt.ylabel('Variances')
```

```python
fig , (ax1,ax2) = plt.subplots(1,2, figsize=(15,5))
# Left plot
ax1.plot(pca2.explained_variance_ratio_, '-o')
ax1.set_ylabel('Proportion of Variance Explained')
ax1.set_ylim(ymin=-0.01)

# Right plot
ax2.plot(np.cumsum(pca2.explained_variance_ratio_), '-ro')
ax2.set_ylabel('Cumulative Proportion of Variance Explained')
ax2.set_ylim(ymax=1.05)
for ax in fig.axes:
    ax.set_xlabel('Principal Component')
    ax.set_xlim(-1,65)
```
